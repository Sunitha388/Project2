```python
import mysql.connector
import pandas as pd
from datetime import datetime

# Connect to the database
connection = mysql.connector.connect(
    host="gateway01.ap-southeast-1.prod.aws.tidbcloud.com",
    port=4000,
    user="2LwupNhEckogWnm.root",
    password="oHH2C3xTDDLFymia",
    database="project2_Top_Final_stock"
)

cursor = connection.cursor()




# Create table
create_table_query = """
CREATE TABLE IF NOT EXISTS monthly_gainers_losers (
    month_year DATE,
    ticker VARCHAR(20),
    daily_return FLOAT
);
"""

cursor.execute(create_table_query)
connection.commit()

print("Table 'monthly_gainers_losers' created successfully.")

#cursor.close()
#connection.close()

# Load the CSV file
df = pd.read_csv(r'C:\Users\LENOVO\Desktop\Newfolder\project2\output\monthly_retn.csv')

# Insert query
insert_query = """
INSERT INTO monthly_gainers_losers (month_year, ticker, daily_return)
VALUES (%s, %s, %s)
"""

# Insert rows into the table
for _, row in df.iterrows():
    try:
        # Convert 'MonthYear' string like "Apr-2024" to date format (e.g., 2024-04-01)
        month_date = datetime.strptime(row['MonthYear'], '%b-%Y').date().replace(day=1)
        cursor.execute(insert_query, (month_date, row['Ticker'], row['Daily_Return']))
    except Exception as e:
        print(f" Error inserting row: {row.to_dict()} → {e}")

connection.commit()
print(" Data inserted into 'monthly_gainers_losers' table successfully.")

cursor.close()
connection.close()
```

```python
import mysql.connector
import pandas as pd
import numpy as np

# Connect to MySQL (TiDB)
connection = mysql.connector.connect(
    host="gateway01.ap-southeast-1.prod.aws.tidbcloud.com",
    port=4000,
    user="2LwupNhEckogWnm.root",
    password="oHH2C3xTDDLFymia",
    database="project2_Top_Final_stock"
)
cursor = connection.cursor()

# Create table if it doesn't exist
create_table_query = """
CREATE TABLE IF NOT EXISTS cumm_top5 (
    Ticker VARCHAR(50),
    close FLOAT,
    date DATE,
    high FLOAT,
    low FLOAT,
    month VARCHAR(20),
    open FLOAT,
    volume BIGINT,
    prev_close FLOAT,
    daily_return FLOAT,
    cumulative_return FLOAT
)
"""  
cursor.execute(create_table_query)
print(" Table 'cumm_top5' created or already exists.")

# Load CSV
df = pd.read_csv('cumulative_returnsdetails1.csv')

# Data cleanup
df['date'] = pd.to_datetime(df['date']).dt.date
df['prev_close'] = df.groupby('Ticker')['close'].shift(1)
df['daily_return'] = (df['close'] - df['prev_close']) / df['prev_close']

def calculate_cumulative_return(x):
    return (1 + x).cumprod() - 1

df['cumulative_return'] = df.groupby('Ticker')['daily_return'].transform(calculate_cumulative_return)

# Convert columns to numeric
cols_to_convert = ['close', 'high', 'low', 'open', 'volume', 'prev_close', 'daily_return', 'cumulative_return']
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Replace null-like values with None
df.replace(["nan", "NaN", np.nan, pd.NaT, pd.NA], None, inplace=True)
df = df.where(pd.notna(df), None)

# Debugging info (optional)
print("\n Data Preview:")
print(df.head(10))
print("\n🔍 Missing Values Count:")
print(df.isnull().sum())

# Prepare data for insertion
data_to_insert = [
    (
        row['Ticker'],
        row['close'],
        row['date'],
        row['high'],
        row['low'],
        row['month'],
        row['open'],
        row['volume'],
        row['prev_close'],
        row['daily_return'],
        row['cumulative_return']
    )
    for _, row in df.iterrows()
]

# Insert block
insert_query = """
INSERT INTO cumm_top5 (
    Ticker, close, date, high, low, month, open, volume,
    prev_close, daily_return, cumulative_return
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

try:
    cursor.executemany(insert_query, data_to_insert)
    connection.commit()
    print(" Data inserted successfully into 'cumm_top5'.")
except Exception as e:
    print(" Error during data insertion:", e)

# Close connection
cursor.close()
connection.close()
print("🔒 MySQL connection closed.")
```
